```python

import nltk
import random
import string

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Knowledge Base
responses = {
    "greeting": [
        "Hello! How can I help you?",
        "Hi there!",
        "Hey! Nice to meet you."
    ],

    "goodbye": [
        "Goodbye!",
        "See you later!",
        "Have a great day!"
    ],

    "name": [
        "I am an AI Chatbot developed using NLP and Python."
    ],

    "course": [
        "I can answer questions related to programming, AI, NLP, and general topics."
    ],

    "python": [
        "Python is a high-level programming language used in AI, Data Science, and Web Development."
    ],

    "ai": [
        "Artificial Intelligence enables machines to mimic human intelligence."
    ],

    "nlp": [
        "Natural Language Processing helps computers understand human language."
    ],

    "default": [
        "Sorry, I don't understand that.",
        "Can you please rephrase your question?",
        "I'm still learning. Try asking something else."
    ]
}

# Stop words
stop_words = set(stopwords.words('english'))

# Intent Detection
def detect_intent(user_input):

    user_input = user_input.lower()

    tokens = word_tokenize(user_input)

    filtered_tokens = [
        word for word in tokens
        if word not in stop_words and word not in string.punctuation
    ]

    if any(word in filtered_tokens for word in ['hello', 'hi', 'hey']):
        return "greeting"

    elif any(word in filtered_tokens for word in ['bye', 'goodbye']):
        return "goodbye"

    elif any(word in filtered_tokens for word in ['name']):
        return "name"

    elif any(word in filtered_tokens for word in ['course', 'subjects', 'learn']):
        return "course"

    elif any(word in filtered_tokens for word in ['python']):
        return "python"

    elif any(word in filtered_tokens for word in ['ai', 'artificial']):
        return "ai"

    elif any(word in filtered_tokens for word in ['nlp']):
        return "nlp"

    return "default"

# Chat Function
def chatbot():

    print("=" * 50)
    print(" AI CHATBOT WITH NLP ")
    print(" Type 'exit' to quit ")
    print("=" * 50)

    while True:

        user_input = input("\nYou: ")

        if user_input.lower() == "exit":
            print("\nBot: Goodbye!")
            break

        intent = detect_intent(user_input)

        response = random.choice(responses[intent])

        print("Bot:", response)

# Run chatbot
if __name__ == "__main__":
    chatbot()
```
